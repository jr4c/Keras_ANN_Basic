```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow.contrib.keras as k
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("data/train.csv")
```

```python
categorical = [c for c in data.columns if c.find("_cat") > 0]
binary = [c for c in data.columns if c.find("_bin") > 0]
numerical = [c for c in data.columns if not c in categorical+binary]
```

```python
data.describe()
```

```python
data[categorical+binary].describe() 
```

```python
n = data[numerical].copy()
```

```python
n.drop(['id'],inplace=True,axis=1)
```

```python
y_train = np.array(n.loc[:,'target':'target'].copy())
x_train = np.array(n.drop(['target'],axis=1).copy())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train,y_train,test_size=0.3)
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
N_FEATURES = 26
```

```python
inputs = k.layers.Input(shape=[N_FEATURES])
net = inputs

net = k.layers.Dense(64,use_bias=False,activation=None)(net)
net = k.layers.BatchNormalization()(net)
net = k.layers.Activation("relu")(net)

net = k.layers.Dense(48,use_bias=False,activation=None)(net)
net = k.layers.BatchNormalization()(net)
net = k.layers.Activation("relu")(net)

net = k.layers.Dense(36,use_bias=False,activation=None)(net)
net = k.layers.BatchNormalization()(net)
net = k.layers.Activation("relu")(net)

net = k.layers.Dropout(0.4)(net)

net = k.layers.Dense(1,activation="sigmoid")(net)

outputs = net
```

```python
model = k.models.Model(inputs,outputs)
```

```python
o = k.optimizers.Adam(lr=0.0001)
l = k.losses.binary_crossentropy
m = ["acc"]
model.compile(optimizer=o,loss=l,metrics=m)
```

```python
model.fit(x_train,y_train,batch_size=2500,epochs=10,validation_split=0.2)
```

```python
print(model.evaluate(scaler.transform(x_test),y_test))
print(model.metrics_names)
```

```python
test_data = pd.read_csv("data/test.csv")
ids = test_data.loc[:,"id":"id"]
test_data = test_data[numerical[2:]]
test_data = scaler.transform(np.array(test_data))
```

```python
y_hat = model.predict(test_data,batch_size=2500)
```

```python
csv_pd = pd.DataFrame({'id': [i for ii in np.array(ids) for i in ii], 'target': [i for ii in np.round(y_hat,4) for i in ii]})
```

```python
csv_pd.to_csv("data/result.csv",index=False)
```
